```python
import models
import data
import numpy as np
import targets
import particle_filters
import matplotlib.pyplot as plt
```

```python
testModel = models.lgss()
testModel.par[0] = 0.7
testModel.par[1] = 0.2
testModel.par[2] = 1
testModel.par[3] = 1
testModel.updateMaxPdf()
```

```python
numParticles = 500
backwardDraws = 2
target = targets.testFun()
data = data.data()
data.generate(testModel, 1000)
```

```python
pf = particle_filters.PaRIS()
pf.setup(testModel, numParticles, backwardDraws, data, target)
pf.run()
```

```python
print(pf.filtMean)
print(pf.filtMean.shape)
```
